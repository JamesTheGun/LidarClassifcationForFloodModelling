```python
import numpy as np
import matplotlib.pyplot as plt

def visualize_tensor(tensor, nodata=-9999.0):
    image_np = tensor.numpy()

    # Drop to 3 channels if needed
    if image_np.shape[2] > 3:
        image_np = image_np[:, :, :3]

    # Mask NoData
    mask = image_np == nodata
    image_np = np.where(mask, np.nan, image_np)

    # Minâ€“max normalise per-channel
    vmin = np.nanmin(image_np, axis=(0, 1), keepdims=True)
    vmax = np.nanmax(image_np, axis=(0, 1), keepdims=True)
    image_np = (image_np - vmin) / (vmax - vmin + 1e-8)

    plt.imshow(image_np)
    plt.axis("off")
    plt.show()
    
visualize_tensor(neg_tensor)
```
